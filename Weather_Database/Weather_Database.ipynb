```python
#import libraries 
import random
import numpy as np
import pandas as pd
#Must create random 2,000 random latitudes and longitudes

#latitudes range from -90 to 90 for the entire world
#longitudes range from -180 to 180 for the entire world 
lats=np.random.uniform(-90,90, size=2000)  
lngs= np.random.uniform(-180,180, size=2000)
```

```python
#3 create coordinates
#now that you have two series you need to create touples with 
#you can do that using the zip and list commands
#https://realpython.com/python-zip-function/
coordinates=list(zip(lats,lngs))
print(coordinates)
```

```python
# 4 Get the nearest city using the citipy module.

#first youll have to import citipy module 
from citipy import citipy
#create an empty city series
cities=[]
#using the coordinates create a for loop that will determine the closes cities near those coordinates
for coordinate in coordinates:
    #use the nearest city function along with the city_name property 
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
cities
```

```python
# 5 Perform an API call with the OpenWeatherMaptarget
# Assuming this is for all cities

#import the requests library.
import requests
# Import the API key.
from config import weather_api_key

city_data=[]
#set up a for loop for all the cities

#also do a try except error in case there is a city
for city in cities:
    #set up url and
    url="http://api.openweathermap.org/data/2.5/weather?q="
    #take note that there will be cities with two names, these cities should have a "+"  instead of a " "
    try:
        city=city
        target_url=url + city.replace(" ","+")+"&appid=" + weather_api_key
        city_weatherdata=requests.get(target_url).json()
        #6 retrieve information from API call
        city_lat = city_weatherdata["coord"]["lat"]
        city_lng = city_weatherdata["coord"]["lon"]
        city_max_temp = city_weatherdata["main"]["temp_max"]
        city_humidity = city_weatherdata["main"]["humidity"]
        city_clouds = city_weatherdata["clouds"]["all"]
        city_wind = city_weatherdata["wind"]["speed"]
        city_country = city_weatherdata["sys"]["country"]
        city_weather_description = city_weatherdata["weather"][0]["description"]
    except:
        print(f"{city} Does not have API information or Error")
    #now we must add them to a data frame
    city_data.append({"City":city,
                  "Latitude":city_lat,
                  "Longitude": city_lng,
                  "Max Temp": city_max_temp,
                  "Humidity": city_humidity,
                  "Cloudiness": city_clouds,
                  "Wind Speed": city_wind,"Country":city_country,"Current Description":city_weather_description})
city_data_df=pd.DataFrame(city_data)
```
